```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('user_session.csv')
```

```python
df.head()
```

```python
df.sort_values(['user_id', 'unix_timestamp'], inplace=True)
df['timestamp_pre'] = df.groupby('user_id')['unix_timestamp'].shift(1)
df['time_delta'] = df['unix_timestamp'] - df['timestamp_pre']
```

```python
df['is_new_session'] = 0

df.loc[(df['time_delta'].isna()) | (df['time_delta']>60*30), 'is_new_session'] = 1
```

```python
df['session_count'] = df.groupby('user_id')['is_new_session'].cumsum()
```

```python
result = df[['user_id', 'page', 'unix_timestamp', 'session_count']]
```
